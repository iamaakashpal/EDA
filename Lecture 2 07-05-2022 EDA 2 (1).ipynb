```python
feature_nan = [i for i in dataset.columns if dataset[i].isnull().sum()]
```

```python
for i in feature_nan:
    print('The',i,'has',np.round(dataset[i].isnull().mean()*100,4),'% Missing Value')
```

```python
# Categorical Feature

Categorical_feature = [i for i in dataset.columns if dataset[i].dtypes == 'object']
```

```python
data = dataset.copy()
def replace_cat_feature(dataset,Categorical_feature):
    data[Categorical_feature] = data[Categorical_feature].fillna('Missing')
    return data
```

```python
replace_cat_feature(data,Categorical_feature)
```

```python
# Numerical Feature

numerical_feature = [i for i in dataset.columns if dataset[i].dtypes != 'object']
```

```python
dataset[numerical_feature].isnull().sum()
```

```python
numerical_feature_nan = [i for i in dataset.columns if dataset[i].dtypes != 'object' and dataset[i].isnull().sum()>=1]
```

```python
numerical_feature_nan
```

```python
for i in numerical_feature_nan:
    # since their are outlier we are going to replace with median
    median_value = dataset[i].median()
    # create a new feature to capture outlier
    dataset[i+'Nan'] = np.where(dataset[i].isnull(),1,0)
    dataset[i].fillna(median_value,inplace=True)
```

```python
# Temporial Variable

Temporial_variable = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']
```

```python
Temporial_variable
```

```python
for i in Temporial_variable:
    dataset[i+' Age of House'] = dataset['YrSold'] - dataset[i]
```
